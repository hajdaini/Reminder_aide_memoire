11. **worst texture (pire texture)** : Texture maximale observée.
12. **worst perimeter (pire périmètre)** : Longueur maximale de la frontière de la tumeur.
13. **worst area (pire surface)** : Taille maximale de la tumeur.
14. **worst smoothness (pire lissage)** : Variation maximale des contours de la tumeur.
15. **worst compactness (pire compacité)** : Proximité maximale des cellules tumorales.
16. **worst concavity (pire concavité)** : Courbure maximale de la tumeur.
17. **worst concave points (pire points concaves)** : Nombre maximal de creux dans la tumeur.
18. **worst symmetry (pire symétrie)** : Symétrie maximale de la tumeur.
19. **worst fractal dimension (pire dimension fractale)** : Complexité maximale de la forme de la tumeur.
20. **benign_0__mal_1** : Indique si la tumeur est bénigne (0) => ou maligne (1) => non-cancéreuse.

## Création du modèle

```python
X = df.drop('benign_0__mal_1', axis=1).values
y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print(X_train.shape)
print(X_test.shape)
print(X_train[0])
print(X_test[0])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

model = Sequential()
model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))
# Binary Classification
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam')
```

## Démo d'un modèle sur entraîné

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test))
```

```python
losses = pd.DataFrame(model.history.history)
ax = losses.plot()
ax.set_title('Progression de la Perte du Modèle Pendant l\'Entraînement/Validation')
plt.show()

# ici le modèle est surentraîné ce qui entraîne une perte de précision sur les données de validation. Il faut ajouter un callback pour arrêter l'entraînement avant que le modèle ne soit surentraîné.
```

## Correction du modèle sur entraîné

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

model = Sequential()
model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))
# Binary Classification
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# Création d'un callback pour l'arrêt anticipé de l'entraînement. L'arrêt anticipé est une forme de régularisation 
# utilisée pour éviter le surapprentissage sur l'ensemble d'entraînement. Il fonctionne en surveillant une 
# quantité spécifique (dans ce cas, 'val_loss'), et en arrêtant l'entraînement si cette quantité cesse de 
# s'améliorer après un certain nombre d'époques ('patience'). Cela permet d'économiser du temps et des ressources 
# en évitant de continuer à entraîner un modèle qui ne s'améliore plus.
early_stop = EarlyStopping(
    monitor='val_loss',  # La métrique à surveiller est la perte sur l'ensemble de validation
    mode='min',  # L'entraînement s'arrête lorsque la métrique surveillée a cessé de diminuer
    verbose=1,  # Mode verbeux : 1 = messages de progression
    patience=25  # L'entraînement s'arrête s'il n'y a pas d'amélioration de la perte sur l'ensemble de validation pendant 25 époques consécutives
)

model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test), callbacks=[early_stop])
```

```python
losses = pd.DataFrame(model.history.history)
ax = losses.plot()
ax.set_title('Progression de la Perte du Modèle Pendant l\'Entraînement/Validation')
plt.show()
```

```python
model = Sequential()
model.add(Dense(30, activation='relu'))

# Intérêt du Dropout : prévenir le surajustement en éteignant aléatoirement certains neurones pendant l'entraînement.
# Cela force le réseau à ne pas trop dépendre de certains neurones spécifiques et améliore la robustesse du modèle,
# l'aidant ainsi à mieux généraliser à de nouvelles données.
model.add(Dropout(0.5))  # Dropout de 50% des neurones

model.add(Dense(15, activation='relu'))
model.add(Dropout(0.5))  # Dropout de 50% des neurones

model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test), callbacks=[early_stop])
```

```python
losses = pd.DataFrame(model.history.history)
ax = losses.plot()
ax.set_title('Progression de la Perte du Modèle Pendant l\'Entraînement/Validation')
plt.show()
```

```python
from sklearn.metrics import classification_report, confusion_matrix

predictions = model.predict(X_test)

# rapport de classification
print(classification_report(y_test, predictions.round()))

print(y_test.shape)
print(confusion_matrix(y_test, predictions.round()))
# explication plus bas
```

Le rapport de classification fournit une analyse détaillée des performances du modèle :

- **Precision (Précision) :** Ratio des vrais positifs sur la somme des vrais positifs et des faux positifs. Ex. : Précision de 0.95 pour la classe 0 indique que 95% des prédictions positives étaient correctes.

- **Recall (Rappel) :** Ratio des vrais positifs sur la somme des vrais positifs et des faux négatifs. Ex. : Recall de 0.98 pour la classe 0 signifie que 98% des instances réelles ont été correctement identifiées.

- **F1-score :** Moyenne harmonique de la précision et du rappel. Un F1-score élevé indique une bonne performance. Ex. : F1-score de 0.96 pour la classe 0.

- **Support :** Nombre d'occurrences de chaque classe dans l'ensemble de test. Ex. : 55 instances de la classe 0, 88 instances de la classe 1.

- **Accuracy (Précision Globale) :** Ratio des prédictions correctes sur le total des prédictions. Ex. : Accuracy de 0.97 indique que le modèle a correctement prédit 97% des instances.

- **Macro avg :** Moyenne des métriques pour chaque classe, sans tenir compte de l'imbalance de classe. Moyenne macro de 0.97.

- **Weighted avg :** Moyenne pondérée des métriques, tenant compte de l'imbalance de classe. Moyenne pondérée de 0.97.

En résumé, le modèle semble avoir une très bonne performance, avec des scores élevés pour la précision, le rappel et le F1-score pour les deux classes.
