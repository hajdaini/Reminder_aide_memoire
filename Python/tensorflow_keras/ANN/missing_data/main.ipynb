**TÂCHE : Il existe de nombreuses façons de traiter ces données manquantes. Nous pourrions essayer de construire un modèle simple pour les remplir, comme un modèle linéaire. Nous pourrions simplement les remplir en fonction de la moyenne des autres colonnes, ou nous pourrions même regrouper les colonnes en catégories, puis définir NaN comme sa propre catégorie. Il n'y a pas d'approche à 100 % correcte ! Examinons les autres colonnes pour voir laquelle est la plus fortement corrélée avec mort_acc.**

```python
df.corr(numeric_only=True)["mort_acc"].sort_values()
```

**TÂCHE : On dirait que la caractéristique "total_acc" (compte total) est corrélée avec "mort_acc" (compte hypothécaire), cela a du sens ! Essayons cette approche fillna(). Nous allons regrouper le dataframe par total_acc et calculer la valeur moyenne de mort_acc par entrée total_acc. Pour obtenir le résultat ci-dessous :**

```python
total_acc_avg = df.groupby('total_acc').mean(numeric_only=True)["mort_acc"]
total_acc_avg
```

> Et maintenant, je peux voir la valeur moyenne du "mort_acc" (compte hypothécaire) par regroupement de compte total. Donc, ici, je peux voir que la moyenne des comptes hypothécaires, si vous avez seulement deux comptes au total, sera de zéro. <br><br>Il est donc logique que si j'ai une ligne où je manque l'information sur le compte hypothécaire, j'utiliserai cette série comme une recherche et je remplacerai cette valeur manquante du compte hypothécaire en fonction de la valeur du compte total que je sais ne pas être manquante, et je la remplirai avec la valeur moyenne.


```python
def fill_mort_acc(total_acc, mort_acc):
    if np.isnan(mort_acc):
        return total_acc_avg[total_acc]
    else:
        return mort_acc
```

```python
df['mort_acc'] = df.apply(lambda x: fill_mort_acc(x['total_acc'], x['mort_acc']), axis=1)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df["mort_acc"].value_counts()
```

# Création du Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

**TÂCHE : Construisez un modèle séquentiel qui sera entraîné sur les données. Vous avez des options illimitées ici, mais voici ce que la solution utilise : un modèle qui va de 78 --> 39 --> 19 --> 1 neurone de sortie. FACULTATIF : Explorez l'ajout de [couches Dropout](https://keras.io/layers/core/) [1](https://en.wikipedia.org/wiki/Dropout_(neural_networks)) [2](https://towardsdatascience.com/machine-learning-part-20-dropout-keras-layers-explained-8c9f6dc4c9ab)**

```python
X_train.shape
```

```python
model = Sequential()

# input layer
model.add(Dense(78,  activation='relu'))
model.add(Dropout(0.4))

# hidden layer
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.4))

# hidden layer
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.4))

# output layer
model.add(Dense(units=1,activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```
