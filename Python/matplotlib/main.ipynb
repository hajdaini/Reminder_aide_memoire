```python
housing = pd.DataFrame({'rooms':[1,1,2,2,2,3,3,3],
                       'price':[100,120,190,200,230,310,330,305]})
housing
```

```python
#scatter = point | plot = points  
plt.scatter(housing["rooms"], housing["price"])
plt.xlim(housing['rooms'].min(), housing['rooms'].max()) # min max des axes x
plt.ylim(housing['price'].min(), housing['price'].max())
plt.title("Rooms and prices")
plt.xlabel("Rooms")
plt.ylabel("Rooms"); # ";"" pour ne pas afficher le print 
```
